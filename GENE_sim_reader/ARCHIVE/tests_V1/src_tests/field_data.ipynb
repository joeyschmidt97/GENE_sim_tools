```python
import os

from GENE_sim_tools.GENE_sim_reader.src.utils.find_buried_filetypes import find_buried_filetypes
from GENE_sim_tools.GENE_sim_reader.src.dict_field_data import field_filepath_to_dict

from GENE_sim_tools.GENE_sim_reader.src.criteria_code.criteria_parser import criteria_parser
```

```python
base_path_132588_q_4 = '/pscratch/sd/j/joeschm/NSXTU_discharges/132588/r_0.736_q=4_MTM_mode'

# test_path = os.path.join(base_path_132588_q_4, 'kymin_scan', 'scanfiles0001')
test_path = os.path.join(base_path_132588_q_4, 'convergence_check', 'kymin_scan_nz0_256', 'scanfiles0000', 'field_0001')
```

```python
crit_str = 'time==131.35920000098082'
time_crit = criteria_parser(crit_str)

field_dict = field_filepath_to_dict(test_path, field_quantities=['field_apar'], time_criteria=time_crit)
```

```python
print(field_dict['field_apar'][0])
```

```python
import numpy as np
import matplotlib.pyplot as plt



# Convert string to list of complex numbers
complex_numbers = np.array([complex(num) for num in field_dict['field_apar'][0]])

# Extract real parts
real_parts = complex_numbers.real
imag_parts = complex_numbers.imag

# Plotting the real part of the complex numbers
plt.figure(figsize=(10, 6))
plt.plot(real_parts, label='Real Part')
plt.plot(imag_parts, label='Imag Part')
plt.title('Plot of the Real Part of Complex Numbers')
plt.xlabel('Index')
plt.ylabel('Real Value')
plt.legend()
plt.grid(True)
plt.show()
```
